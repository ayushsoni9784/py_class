```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(req.text, 'lxml')
```

```python
events = soup.find('ol', {'class':'list-recent-jobs'} )
```

```python
events.find_all('li')
```

```python
des = []
for e in events.find_all('h2'):
    des.append(e.find('a').text)
des
```

```python
time = []
for e in events.find_all('li'):
    time.append(e.find('span', {'class':'listing-posted'}).find('time').text)
time
```

```python
location = []
for e in events.find_all('li'):
    location.append(e.find('span', {'class':'listing-location'}).find('a').text)
location
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['Designation'] = des
df['Time'] = time
df['Location'] = location
```
